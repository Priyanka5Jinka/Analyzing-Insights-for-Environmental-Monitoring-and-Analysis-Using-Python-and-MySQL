```python
import pandas as pd
```

```python
#Task1
#Loading the data

df = pd.read_csv('iot_telemetry_data.csv',low_memory=False)
df
```

```python
#Task 2: Renaming the Columns
new_col = {'ts':'timestamp','device':'device_id','co':'carbon_monoxide','lpg':'liquefied_petroleum_gas','temp':'temperature'}
df = df.rename(columns=new_col)
df
```

```python
#Task 3: check for null values
null_values = df.isnull().sum()
null_values
```

```python
df.duplicated().sum()
```

```python
#Task4 :Removing Duplicates
df.drop_duplicates(inplace=True)
df
```

```python
df.duplicated().sum()
```

```python
#Task 5:Handling Missing Values:
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
#Task 6:Data Type Conversion:
df.info()
```

```python
dicti = { 'timestamp' : 'datetime64[ns]',
         'light' : 'bool',
         'motion':'bool',
         'smoke':'bool'
    }
df = df.astype(dicti)
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df.timestamp)
df.info()
```

```python
df['light'] = df.light.astype('bool')
df.info()
```

```python
df['motion'] = df.motion.astype('bool')
df['smoke'] = df.smoke.astype('bool')

df.info()
```

```python
df
```

```python
#Task 7: Export the cleaned dataset to "cleaned_environemnt.csv"
df = df.to_csv('cleaned_environment.csv',index=False)
```
